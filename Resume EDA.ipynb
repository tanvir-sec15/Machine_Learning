## Testing unique values

```python
print(df['quality'].unique())

## that means quality কলামে শুধু এই ভ্যালুগুলো পাওয়া গেছে। 
# (এদের ডুপ্লিকেট গুলো ইগ্নোর করা হয়েছে )
```

```python
check_uniq = df['quality'].value_counts()
check_uniq

##মানে dataset-এ quality কলামে প্রতিটি রেটিং কতবার এসেছে
# যেমন 5 এসেছে 681 বার, 6 এসেছে 638 বার ইত্যাদি।
```

```python
plt.bar(check_uniq.index,check_uniq.values)
plt.xlabel("Quality")
plt.ylabel("Count")
plt.title("Quality vs Count")
plt.show()

##quality_counts.index → unique quality values (যেমন 3,4,5,6,7,8)
##quality_counts.values → প্রতিটি quality কতবার এসেছে (count)
            
```

```python
df.quality.value_counts().plot(kind='bar')
plt.xlabel("Quality")
plt.ylabel("Count")
plt.title("Quality vs Count")
plt.show()
```

### Correlation

```python
import seaborn as sns
df.corr().head(3)

##Correlation মানে হলো, দুইটা ভেরিয়েবল একে অপরের সাথে কিভাবে সম্পর্কিত।
#A বাড়লে B একইভাবে বাড়ে, A কমলে B কমে (Positive Correlation)    
```

### Heatmap

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.figure(figsize=(10,6))
plt.title("Correlation Heatmap")

##annot=True → প্রতিটি cell-এ মান দেখাবে
```

### Histplot

```python
for columns in df.columns:
    sns.histplot(df[columns])
```

```python
for col in df.select_dtypes(include='number').columns:
    sns.histplot(df[col],bins=10,kde = True,color='blue')
    plt.title(f'Histogram of {col}')

 ## select_dtypes() function দিয়ে আমরা specific data type নির্বাচন করতে পারি
 # যেমন এখানে শুধু numerical column গুলো নির্বাচন করেছি।  
 # kde=True → এটা histogram এর সাথে একটা smooth curve যোগ করে যা distribution বুঝতে সাহায্য করে 
 # Bins → X-axis-কে ভাগ করার সংখ্যা (intervals)
```

```python
sns.histplot(df['alcohol'],bins=10,kde = True,color='blue')
```
