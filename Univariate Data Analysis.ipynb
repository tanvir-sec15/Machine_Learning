```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Univariate Analysis 
##### আমরা একেকটা কলামকে কতভাবে এনালাইসিস করতে পারি সেটাই Univariate analysis

```python
df=pd.read_csv("D:\Machine_Learning\Titanic-Dataset.csv")
df.sample(1)
```

### 1.Categorical Data Analysis 
##### ( Taking single Column to analyse )

### a. Countplot 

```python
df['Survived'].value_counts()
```

##### using Seaborn visualizsation

```python
sns.countplot(x='Survived',data=df)   ## Seaborn Countplot visual representation এর মাধ্যমে 
```

#### using Matplotlib visualization

```python
df.value_counts('Embarked').plot(kind='bar')
```

### b. using PieChart

```python
df.value_counts('Embarked').plot(kind='pie',explode=[0,0.1,0],autopct='%1.1f%%')    ##autopct='%1.1f%%' pie chart এর মধ্যে percentage দেখাবে

##explode pie chart এর মধ্যে একটা অংশ আলাদা করে দেখাবে ( 0.1 মানে 10% আলাদা করে দেখাবে, 0 মানে আলাদা করবে না)
```

### 2.Numerical Data

##### a.Using Histogram

```python
## Numerical data এর distribution বুঝার জন্য Histogram plot

plt.hist(df['Age'],bins=20,edgecolor='black')  ## bins = number of bars
plt.xlabel('Age')
plt.ylabel('Count') 
plt.title('Age Distribution')
plt.show()
```

##### b. Distplot

```python
sns.displot(df['Age'],kde=True,color='Red',edgecolor='Black')  ## kde=True distribution line দেখাবে
plt.xlabel('Age')
plt.ylabel('Density')

''' এইটা Skewness বুঝার জন্য খুবই important । যদি distribution এর left side বেশি হয় তাহলে left skewed (negative skewed), 
যদি distribution এর right side বেশি হয় তাহলে right skewed (positive skewed)
    যদি distribution দুই পাশ সমান হয় তাহলে normal distribution
'''

## Shows probability density function by default ,কোনো Quantity এর distribution এ থাকার সম্ভাবনা কতটুকু সেটা বুঝায় , y axis এ দেখায় density
```

```python
df['Age'].skew()  ##ভ্যাল্যু 0.389 অর্থাৎ right skewed ## যদি ভ্যাল্যু 0 এর কাছাকাছি হয় তাহলে normal distribution
```

#### c. Boxplot

```python
sns.boxplot(x=df['Pclass'],y=df['Age'],color='lightblue')   ## vertical boxplot

## Boxplot এর মাধ্যমে কোন variable এর distribution, median, quartile, outlier ইত্যাদি সম্পর্কে ধারণা পাওয়া যায়
## Range এর মধ্যে data কতটুকু ছড়িয়ে আছে সেটা বুঝতে পারি
## Range এর বাইরে যেগুলো আছে সেগুলোকে outlier ধরা হয় (উদ্ভট মান)
```
